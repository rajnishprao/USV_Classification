```python
import numpy as np
from matplotlib import pyplot as plt
from scipy.io import wavfile
import scipy.signal as signal

%matplotlib inline
```

```python
# Read the wav file (mono)

samplingFrequency, signalData = wavfile.read('sample for spectrograms.wav')
```

```python
plt.specgram(signalData,Fs=samplingFrequency)
plt.xlabel('Time')
plt.ylabel('Frequency')
plt.xlim(110, 115)
plt.ylim(20000, 100000)
```

```python
# increasing fft resolution to 1024 as in paper, and color to inverted gray scale
plt.specgram(signalData,Fs=samplingFrequency,NFFT=1024,cmap='gray_r')#,vmin=0, vmax=255)
plt.xlabel('Time')
plt.ylabel('Frequency')
plt.xlim(110, 111)
plt.ylim(20000, 100000)
```

```python
# checking out interpolation function for cleaner spectrograms: marginally better
plt.specgram(signalData,Fs=samplingFrequency,NFFT=1024,cmap='gray_r', interpolation=None)
plt.xlabel('Time')
plt.ylabel('Frequency')
plt.xlim(110, 111)
plt.ylim(20000, 100000)
```

```python
# adding vmin/vamx, not sure this helps
plt.specgram(signalData,Fs=samplingFrequency,NFFT=1024,cmap='gray_r',interpolation=None,vmin=0, vmax=255)
plt.xlabel('Time')
plt.ylabel('Frequency')
plt.xlim(110, 111)
plt.ylim(20000, 100000)
```
